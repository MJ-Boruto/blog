## 1. 表達與傾聽
- 表達：試著先組織自己的語言(提前準備，包括大綱、表達方式)、邏輯清晰、不過度誇大、舉例、具體形容
- 傾聽：記重點、用複述來確認、搞清楚對方的意思、3F傾聽模型
    - 傾聽事實(Fact)
    - 傾聽感受(Feeling)
    - 傾聽意圖(Focus)
    - 核對彼此的3F並達成共識

### 如何搞砸溝通
說大話、貶低對方、散發負能量、扯一些別人不關心的
### 如何和陌生人尬聊
從對方熟悉的事物開始，接著開始發揮聯想力。可以注意的有：熱情點、主動點、不要糾結太多
### 如何即興談話
- 領獎或聚會時一對多：過去+現在+未來
- 會議討論：現象、原因、短期處理方法、長期處理方法

## 2. 求職(一)：準備階段
### 該做的準備
研究職位、公司、不要空手去面試(帶履歷、筆)、外在形象、面試常問的問題、模擬面試、心理準備
> Tip: 多做準備才不容易緊張

#### 替公司想
所應徵的職位最看重的能力和素質、你能為公司創造什麼價值、高機率只買單你的成功經驗
#### 替人資想
履歷不必華麗只求**清晰傳達資訊**，有三個原則可以參考：
1. 目標原則：內容圍繞職位
2. 簡單原則：刪掉無關項目
3. 對應原則：履歷頁數e.g.畢業生一頁、社會人士兩頁、高層人員多頁。

### 履歷基本框架
1. 個人資訊：姓名、性別、出生年月、居住地、年資、聯絡方式(電話、email一定要寫對)等
2. 自傳：**替招聘人員想**：用最簡潔、突出、一目了然的方式展現自己和職位相符的特徵
    1. 學經歷、業績、專案成果、可獨當一面或帶團隊的工作領域
    2. 優勢：相對同類求職者的特殊才能、資源或經歷。
    3. 願景：展現對應徵公司、職位的渴望
3. 應徵條件：求職目標、期待薪資與工作狀態等
    1. 目標不明確(同時應徵不同部門)會讓別人覺得你不夠穩定
    2. 研究職位的薪資水準再依據個人實際狀況定個合理的範圍(自身價值評估也是個重要的能力)
4. 工作經驗(履歷的重點)：求職空窗期一定要解釋原因
    1. 公司介紹
    2. 擔任職位
    3. 主要職責(用和工作有關的動詞：負責完成、協助完成、組織等)
    4. 離職原因
    5. 業績(用數據)
5. 專案經驗：包括專案中擔任的角色、職責和結果
    1. 和工作經驗比起來，專案經驗更能凸顯你的專業技能水準，因此管理、技術、諮詢類職位更注重於此
    2. 專案名稱+時間+簡介、個人職責、主要業績。
6. 教育經歷：填最高學歷就好
7. 證書與培訓經歷：**與工作有關**的證書與培訓(展現你的學習意願與潛力，是你喜歡學習的證據)
8. 語言能力

## 3. 求職(二)：面試階段
### 自我介紹
1. 角度原則：講面試官想聽的
2. 重點原則：講重點、刪掉和職位無關的話
3. 職業化原則：使用職場語言、太口語化可能會顯得你不專業和隨便

#### 自介具體內容
- 一分鐘版本
    1. 個人資訊、學習經歷
    2. 相關經歷、業績、優勢能力
    3. 表達願景
- 三分鐘版本
    1. 個人資訊、學習經歷
    2. 相關工作經歷、專案、業績
    3. 能力與優勢
    4. 個人興趣、特長，描述工作以外的你，但最好還是能展現職位相關的人格特質
    5. 表達願景，展現對應徵公司、職位的渴望

### 建議
- 熟知履歷的每項內容以及自己值得被錄取的點有哪些
- 講自己的能力時接著說能為企業創造什麼價值
- 重點是面試官想透過問題知道什麼，而不是字面上的意思，這樣才不會離題
- 最好舉過去的實際例子比較有說服力
- 句數多的話可以考慮結構化，例如條列重點(第一...第二...第三...)
- 成功經驗不夠的話，就多利用業餘時間學習、接受培訓

### 面試流程
- 初步篩選：履歷審核、專業考試(研發、財務、人資等職位尤其看重)、性格測試
- 初試：辨別履歷描述的真假、是否符合公司文化
- 複試：一般為專業面試
- 終試：全面考核

|職位|初試人員|複試人員|終試人員
|:-:|:-:|:-:|:-:
|一般職員|人資專員|人資主管+部門主管|部門經理
|主管    |人資專員|人資經理+部門經理|直屬上司
|中高階主管、研發等特殊人才|人資專員/人資主管|人資經理+部門直屬上司|總經理/董事長
> Note: 面試人員因企業規模及具體情況有所不同，複試、終試也可能合併為多對一面試。

### 提問
- 展現你對職位的興趣
- 通用問題：部門組織結構、培訓機會、自己將在公司中的定位、交接事項目前進度(趁機談談你的想法)、希望我來解決什麼問題(趁機銷售自己)
- 一般職位問和職位相關的問題就好，例如：企業文化、職責與該職位目前的挑戰、部門組織結構
- 管理職位問和公司相關的問題甚至是公司時事，例如：公司組織結構、公司產業別的發展趨勢
- 適量就好，大概三到五個
- 別問薪資問題，有興趣會主動跟你談
- 別問傻問題，例如：問自己表現得怎麼樣、能否預支薪水、非上市公司的營利問題和未來策略

## 4. 工作
### 與上司
有自信、察言觀色、主動彙報、不過度隱藏弱點
#### 彙報
- 過程彙報：現況與進度+**困難與問題(彙報主因)**+想到的解法 + 主管意見 = 未來計畫
- 結果彙報：告知結果

#### 方案審核
- **先搞清楚上司的喜好與著重點**
- 溝通公式：闡述+優劣+建議+結果
- 最好有第二方案
- 蒐集好相關資料(以推薦第一方案)

### 與同事
製造雙贏、關心與體諒、別等有事才找同事、有事要主動催
### 與下屬
ESA溝通模型：鼓勵(Encourage)、支持(Support)、問責(Accountabiliy)
## 5. 人性
- 人總有虛榮心的需求，禮多人不怪
- 展現專業度才容易得到認可與尊重
- 人總是自私的，別期望別人先幫你
- 有交集點時更容易打開心房
- 表現誠懇為上策，人對小聰明很敏感

> Tip: 溝通能力差可能只因為沒行動力而已，沒行動力的話可以試著想想做與不做的差別(短期與長期)
