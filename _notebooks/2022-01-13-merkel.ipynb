## 編年
|時間|事件
|:|:-
|1950|舒曼宣言，呼籲成立歐洲煤鋼共同體
|1954|梅克爾出生於西德漢堡，隨後移居至東德
|1961|修建柏林圍牆
|1989|11月9號柏林圍牆倒塌，梅克爾加入民主覺醒黨，該黨部不久後與基民黨合併，梅克爾於東德第一次民主選舉後擔任新政府副發言人
|1990|兩德統一，梅克爾擔任婦女青年部長，普丁逃離格勒斯登。
|1994|梅克爾擔任環境和核能安全部長
|1995|第一屆聯合國氣候會議，簽訂《柏林授權書》(Berlin Mandate)，要求各國設定有法律約束力的目標及時間表，促成兩年後的《京都議定書》(Tokyo Protocol)
|1999|歐元取代馬克後
|2003|德法反對美國攻打伊拉克
|2004|歐盟東擴，由15國展至25國
|2005|五六月，法荷公投反對歐盟憲法條約(Constitutional Treaty)
|    |十一月，梅克爾首次組閣，擔任德國總理 
|2007|里斯本條約
|2008|以色列建國60週年，梅克爾至以色列國會演講致歉大屠殺
|2009|金融海嘯
|2010|歐債危機，希臘波及歐元區
|2011|阿拉伯之春(地中海南岸)衝擊回教世界，突尼西亞、埃及、葉門政府垮臺
|    |美法等國出兵干涉利比亞(格達費)
|    |利比亞、敘利亞內戰開始
|2012|歐盟建立銀行聯盟
|2014|克里米亞危機
|    |梅克爾聯手法國、波蘭，說服普亭建立明斯克議定書(Minsk Protocol)防堵烏克蘭危機擴大
|2015|中東北非難民湧入歐洲
|    |開放難民入德
|2017|建立獨立於北約的聯合防衛
|2018|梅克爾辭去黨魁，為了專心政務
|2020|一月底，英國脫歐
|2021|首次發行歐盟債券
|    |十二月，梅克爾卸任總理

> Note: 德式民主：選舉經費不是天價，選舉經費有嚴格的限制與控管，降低從政的財力門檻。聯邦制，加上多黨又必須組織聯合政府，「協商」成為了德國政治的特色與文化以及領袖的必備能力。
